## Automate Writing the connection message

```python
import pandas as pd
import numpy as np

data = pd.read_csv('./Data/data.csv')
data.head()# split the name into first and last name for the first name column
greet_message = "Hi, [firstname]! I am Ken, the co-founder at Esger, an NYC startup. We just raised $250k+ and support from top institutions like Cornell, Microsoft, and ERA. As we are using AI to tackle the heart of sustainability issues, the supply chain, I love to connect with experts in the field (like you!)."

data['first_name'] = data['Name'].apply(lambda x: x.split(' ')[0] if type(x) == str else np.nan)
data['greet_message'] = data['first_name'].apply(lambda x: greet_message.replace('[firstname]', x) if type(x) == str else np.nan)

data = data[["Name",'By','greet_message']]

data.to_csv('./Data/processed_data.csv', index=False)
```

## Automate Outputing the name of linkedin Contact

```python
import pandas as pd
import numpy as np
import re

by_who = "Ken"

greet_message = "Hi, [firstname]! I am Ken, the co-founder at Esger, an NYC startup. We just raised $250k+ and support from top institutions like Cornell, Microsoft, and ERA. As we are using AI to tackle the heart of sustainability issues, the supply chain, I love to connect with experts in the field (like you!)."

input_string = '''https://www.linkedin.com/in/melissa-mardarie-53364117/ | (1) Melissa Mardarie | LinkedIn
https://www.linkedin.com/in/na-yang-977b7b8/ | (1) NA YANG | LinkedIn
'''


def remove_special_characters_and_numbers(input_str):
    # Remove all characters that are not alphabets
    clean_str = re.sub(r'[^a-zA-Z\s]', '', input_str)
    return clean_str

# Initialize an empty list to store the names
names_list = []
url_list = []
# Split the input string by lines
for line in input_string.split("\n"):
    # Split each line by the '|' character
    parts = line.split(" | ")

    # The name is the second element in the 'parts' list, strip leading and trailing whitespaces
    url = parts[0].strip()
    name = remove_special_characters_and_numbers(parts[1]).strip()
    
    # Append the name to the list
    names_list.append(name)
    url_list.append(url)


output = pd.DataFrame({"Name":names_list, "Who?":by_who, "LinkedIn Profile":url_list})
output['greet_message'] = output['Name'].apply(lambda x: greet_message.replace('[firstname]', x.split(' ')[0]) if type(x) == str else np.nan)

output.to_csv("./Data/outreach_parsed.csv", index=False)
```
